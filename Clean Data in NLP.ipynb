```python
import re
import string

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

```python
for i in range(len(all_positive_tweet)):
    all_positive_tweet[i] = re.sub(r"^RT[\s]+", "", all_positive_tweet[i])
    all_positive_tweet[i] = re.sub(r'https?://[^\s\n\r]+', '', all_positive_tweet[i])
    all_positive_tweet[i] = re.sub(r'#', '', all_positive_tweet[i])
```

```python
all_positive_tweet
```

```python
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True,
                               reduce_len=True)
```

```python
all_positive_tweet_tokenize_test = [None] * len(all_positive_tweet)
for i in range(len(all_positive_tweet)):
  all_positive_tweet_tokenize_test[i] = tokenizer.tokenize(all_positive_tweet[i])
```

```python
all_positive_tweet_tokenize_test
```

```python
all_positive_tweet_tokenize = []
for i in range(len(all_positive_tweet)):
  all_positive_tweet_tokenize.append(tokenizer.tokenize(all_positive_tweet[i]))
```

```python
all_positive_tweet_tokenize
```

# Remove stop words and punctuations

```python
stopwords_english = stopwords.words("english")

print("Stop Words")
print(stopwords_english)

print()
print(string.punctuation)
```

```python
for i in range(len(all_positive_tweet_tokenize)):
    all_positive_tweet_tokenize[i] = [word for word in all_positive_tweet_tokenize[i] if word not in stopwords_english]
```

```python
all_positive_tweet_tokenize
```

```python
import string

all_positive_tweet_tokenize_clean = []

# Iterate over each string in all_positive_tweet_tokenize
for i in range(len(all_positive_tweet_tokenize)): 
    # Remove punctuation from each string and append to all_positive_tweet_tokenize_clean
    all_positive_tweet_tokenize_clean.append([''.join(char for char in s if char not in string.punctuation) for s in all_positive_tweet_tokenize[i]])

# Print the cleaned list
print(all_positive_tweet_tokenize_clean)
```

```python
stemmer = PorterStemmer()

# Create an empty list to store the stems
tweets_stem = []

# Iterate over each list of strings in all_positive_tweet_tokenize_clean
for word_list in all_positive_tweet_tokenize_clean:
    # Iterate over each string in the list
    for word in word_list:
        # Stem the word and append to tweets_stem
        stem_word = stemmer.stem(word)
        tweets_stem.append(stem_word)
```
